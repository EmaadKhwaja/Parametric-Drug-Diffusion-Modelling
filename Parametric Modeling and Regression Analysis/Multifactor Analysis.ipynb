```python
import pandas as pd
import prince
import matplotlib as plt

df= pd.read_csv("/Users/emaad/Google Drive/Research/Graduate School/Ting Xu Rotation/Model Fits/Params Table - Copy.csv")

groups={'Quant':['Permeability (P)', 'Convection Term (X)',
        'Size (nm)']}


df=df.dropna()

Lipid=df[df['Particle Type']=='Lipid']

Polymer=df[df['Particle Type']=='Polymer']
Carbon=df[df['Particle Type']=='Carbon']
Gold=df[df['Particle Type']=='Gold']


mfa=prince.MFA(groups=groups,n_components=2,n_iter=3,copy=True,check_input=True,engine='auto',random_state=42)
mfa=mfa.fit(df)

```

```python
>>> bx = mfa.plot_row_coordinates(
...     Lipid,
...     ax=None,
...     figsize=(10, 10),
...     x_component=0,
...     y_component=1,
        color_labels=Lipid['Tumor'],
...     ellipse_outline=False,
...     ellipse_fill=False,
...     show_points=True,
... )

bx.get_figure().savefig('Polymer - Size.svg')
```
